```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import csv
from matplotlib.pyplot import figure

dates = ['2018', '2019', '2020', '2021']
figure(figsize=(8, 6), dpi=100)

df = pd.read_csv('table1.csv')
x = df['Year'].round(1)
y1 = df['Total']
y2 = df['Death']
plt.plot(dates, y1, label =  "Загальна кількість",  linestyle='--', marker='o')
plt.plot(dates, y2, label = "Смертельні наслідки",  linestyle='--', marker='o')

plt.xlabel('Роки')
plt.ylabel('Кількість осіб, що постраждало')
plt.legend()

plt.show()
%matplotlib inline
```
